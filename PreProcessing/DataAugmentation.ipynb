```
originalImage = cv2.imread(r'/content/drive/MyDrive/Dataset/Dataset_Resized/trash/trash10.jpg')
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(originalImage)
```

```
flipVertical = cv2.flip(originalImage, 0)
flipHorizontal = cv2.flip(originalImage, 1)
flipBoth = cv2.flip(originalImage, -1)
```

```
cv2_imshow(originalImage)
cv2_imshow(flipVertical)
cv2_imshow(flipHorizontal)
cv2_imshow(flipBoth)
```

```
img_path = r"/content/drive/MyDrive/Dataset/Dataset_Resized/trash"
```

```
import os
import glob
img_list = glob.glob(os.path.join(img_path,'*.jpg'))
print(len(img_list))
```

```
print(img_path + "/")
```

```
count = len(img_list)
for img in img_list:
    originalImage = cv2.imread(img)
    flipBoth = cv2.flip(originalImage, -1)
    count = count + 1
    cv2.imwrite(img_path + '/' + 'trash' + str(count) + '.jpg', flipBoth)
    print(img_path + '/' + 'trash' + str(count) + '.jpg')
```
