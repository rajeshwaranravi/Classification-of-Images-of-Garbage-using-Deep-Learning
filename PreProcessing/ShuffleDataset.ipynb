```
import random

for folder in os.listdir(img_dir):
  print(folder)
  img_path = glob.glob(os.path.join(img_dir, folder + '/*.jpg'))
  random.shuffle(img_path)
  num = len(img_path)
  print(num)
  print('Train: ', int(0.8*num))
  train = img_path[:int(0.8*num)]
  print('Validation: ',  int(0.8*0.2*num))
  validation = img_path[int(0.8*num):int(0.8*0.2*num)+int(0.8*num)]
  print('Test: ', (num - int(0.8*num) - int(0.8*0.2*num)))
  test = img_path[int(0.8*0.2*num)+int(0.8*num):(num - int(0.8*num) - int(0.8*0.2*num) + int(0.8*0.2*num)+int(0.8*num))]
  
  os.makedirs(dest_dir + '/Train/' + folder)
  for img in train:
    imgtemp = numpy.array(cv2.imread(img))
    cv2.imwrite(dest_dir + '/Train/' + folder + '/' + '{}.jpg'.format(os.path.split(img)[-1].split('.')[0]),imgtemp)
    print(dest_dir + '/Train/' + folder + '/' + '{}.jpg'.format(os.path.split(img)[-1].split('.')[0]))
  img_list = list()
  img_list = glob.glob(os.path.join(dest_dir + '/Train/' + folder, '*.jpg'))
  print(len(img_list))

  os.makedirs(dest_dir + '/Validation/' + folder)
  for img in validation:
    imgtemp = numpy.array(cv2.imread(img))
    cv2.imwrite(dest_dir + '/Validation/' + folder + '/' + '{}.jpg'.format(os.path.split(img)[-1].split('.')[0]),imgtemp)
    print(dest_dir + '/Validation/' + folder + '/' + '{}.jpg'.format(os.path.split(img)[-1].split('.')[0]))
  img_list = list()
  img_list = glob.glob(os.path.join(dest_dir + '/Validation/' + folder, '*.jpg'))
  print(len(img_list))

  os.makedirs(dest_dir + '/Testing/' + folder)
  for img in test:
    imgtemp = numpy.array(cv2.imread(img))
    cv2.imwrite(dest_dir + '/Testing/' + folder + '/' + '{}.jpg'.format(os.path.split(img)[-1].split('.')[0]),imgtemp)
    print(dest_dir + '/Testing/' + folder + '/' + '{}.jpg'.format(os.path.split(img)[-1].split('.')[0])) 
  img_list = list()
  img_list = glob.glob(os.path.join(dest_dir + '/Testing/' + folder, '*.jpg'))
  print(len(img_list))
```
