```
def resize_img(categories):
    for category in categories:
        img_list = glob.glob(os.path.join(img_dir, category + '/*.jpg'))
        print(img_list)
        print(len(img_list))
        img_dest_dir = os.path.join(dest_dir, category)
        print(img_dest_dir)
        os.chdir(img_dest_dir)
        for img in img_list:
          imgtemp = numpy.array(cv2.imread(img))
          imgtemp = cv2.resize(imgtemp,(512,384))
          cv2.imwrite('{}.jpg'.format(os.path.split(img)[-1].split('.')[0]),imgtemp)
```

```
resize_img(categories)
```
